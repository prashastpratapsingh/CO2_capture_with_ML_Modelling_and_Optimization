Splitting the data into Train and Test with test sample size of 0.2 of train. 
This is done for model training and knowing the accuracy of models

```python

# import packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
				
X= df[['X (Mole Fraction in Amine Mixture)','Total Amine Concentration','Temperature','Partial Pressure of CO2']]
y=df['CO2 loading (mol CO2/mol amine)']

# using the train test split function
X_train, X_test,y_train, y_test = train_test_split(X,y ,
								random_state=104,
								test_size=0.2,
								shuffle=True)

# printing out train and test sets

print('X_train : ')
print(X_train.head())
print('')
print('X_test : ')
print(X_test.head())
print('')
print('y_train : ')
print(y_train.head())
print('')
print('y_test : ')
print(y_test.head())
```

## **Linear Regression**
Linear Regression is a machine learning algorithm based on supervised learning. It performs a regression task. Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x)). Hence, the name is Linear Regression.
 The main mathematical function is 
 Y = a+bX

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
reg.score(X_test, y_test)
```

The accuracy of Linear Regression model obtained is 69.967%

# **Support Vector Machine**
Support Vector Machine (SVM) Regressor is a type of machine learning model that is used for regression analysis. It belongs to a class of models called supervised learning models, where the model is trained on a labeled dataset to predict the value of an output variable based on one or more input variables.

The SVM Regressor works by finding a hyperplane in a high-dimensional space that maximizes the margin between the training data points and the hyperplane. The margin is the distance between the hyperplane and the closest data points. In a regression problem, the goal is to find a hyperplane that best fits the data points and makes accurate predictions for new data.


```python
from sklearn import svm
regr = svm.SVR(kernel='linear')
regr.fit(X_train, y_train)
regr.score(X_test,y_test)
```

The accuracy of SVM model obtained is 47.294%

# **Decision Tree Regressor**
Decision Tree Regression is a type of machine learning model that is used for regression analysis. It is based on the concept of a decision tree, where each internal node of the tree represents a decision based on a feature of the input data, and each leaf node represents a prediction of the output variable.

The Decision Tree Regression algorithm works by recursively splitting the input data into subsets based on the value of a selected feature, such that the subsets become more homogeneous with respect to the output variable. The splitting process continues until a stopping criterion is met, such as a maximum depth of the tree or a minimum number of data points in a leaf node.


```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

The accuracy of Decision Tree model obtained is 76.168%

# **Random Forest**
Random Forest Regression is a type of ensemble learning model that is used for regression analysis. It is based on the concept of a decision tree, where multiple decision trees are trained on different subsets of the input data, and the final prediction is made by averaging the predictions of the individual trees.

The Random Forest Regression algorithm works by randomly selecting a subset of the input data and a subset of the input features for each decision tree. This randomness helps to reduce overfitting and increase the diversity of the individual trees. During training, each decision tree is trained using the same algorithm as the Decision Tree Regression model, where the goal is to minimize the variance of the output variable within each leaf node.


```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(random_state=0)
regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

The accuracy of Random Forest model obtained is 80.99%

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(random_state=0, n_estimators=80, max_depth=14)
regressor.fit(X_train,y_train)
regressor.score(X_test,y_test)
```

The accuracy of Random Forest model obtained after relevant hyperparameter tuning is 81.18%
